```python
def count_vowels(string):
    # Convert the string to lowercase to handle both uppercase and lowercase vowels
    string = string.lower()

    # Initialize a dictionary to store the vowel counts
    vowel_counts = {'a': 0, 'e': 0, 'i': 0, 'o': 0, 'u': 0}

    # Iterate over each character in the string
    for char in string:
        # Check if the character is a vowel
        if char in vowel_counts:
            # Increment the count of the vowel in the dictionary
            vowel_counts[char] += 1

    return vowel_counts

# Test the function
input_string = input("Enter a string: ")
result = count_vowels(input_string)
print("Number of each vowel in the string:")
for vowel, count in result.items():
    print(f"{vowel}: {count}")
```

```python
def count_string_info(string):
    # Count the number of words
    words = string.split()
    word_count = len(words)

    # Count the number of characters
    char_count = len(string)

    # Count the number of spaces
    space_count = string.count(" ")

    return word_count, char_count, space_count

# Test the function
input_string = input("Enter a string: ")
word_count, char_count, space_count = count_string_info(input_string)
print("Number of words:", word_count)
print("Number of characters:", char_count)
print("Number of spaces:", space_count)
```

```python
def remove_punctuation(string):
    # Define a string of punctuation marks
    punctuation = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''

    # Initialize an empty string to store the result
    result = ""

    # Iterate over each character in the string
    for char in string:
        # Check if the character is not a punctuation mark
        if char not in punctuation:
            # Append the character to the result string
            result += char

    return result

# Test the function
input_string = input("Enter a string: ")
result_string = remove_punctuation(input_string)
print("String with punctuation removed:", result_string)
```

```python
def create_matrix(rows, cols):
    matrix = []
    for i in range(rows):
        row = []
        for j in range(cols):
            element = int(input(f"Enter element at position ({i}, {j}): "))
            row.append(element)
        matrix.append(row)
    return matrix

def add_matrices(matrix1, matrix2):
    rows = len(matrix1)
    cols = len(matrix1[0])
    result = []
    for i in range(rows):
        row = []
        for j in range(cols):
            element = matrix1[i][j] + matrix2[i][j]
            row.append(element)
        result.append(row)
    return result

def multiply_matrices(matrix1, matrix2):
    rows1 = len(matrix1)
    cols1 = len(matrix1[0])
    rows2 = len(matrix2)
    cols2 = len(matrix2[0])
    result = []
    for i in range(rows1):
        row = []
        for j in range(cols2):
            element = 0
            for k in range(cols1):
                element += matrix1[i][k] * matrix2[k][j]
            row.append(element)
        result.append(row)
    return result

def transpose_matrix(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    result = []
    for j in range(cols):
        row = []
        for i in range(rows):
            element = matrix[i][j]
            row.append(element)
        result.append(row)
    return result

# Prompt the user to input the dimensions of the matrices
rows = int(input("Enter the number of rows: "))
cols = int(input("Enter the number of columns: "))

# Create the first matrix
print("Enter elements for the first matrix:")
matrix1 = create_matrix(rows, cols)

# Create the second matrix
print("Enter elements for the second matrix:")
matrix2 = create_matrix(rows, cols)

# Perform matrix addition
addition_result = add_matrices(matrix1, matrix2)

# Perform matrix multiplication
multiplication_result = multiply_matrices(matrix1, matrix2)

# Perform matrix transpose
transpose_result = transpose_matrix(matrix1)

# Print the results
print("Matrix Addition:")
for row in addition_result:
    print(row)

print("Matrix Multiplication:")
for row in multiplication_result:
    print(row)

print("Matrix Transpose:")
for row in transpose_result:
    print(row)
```
